# JointPlot 

jointplot() allows you to basically match up two displota for bivariabte data.

```python
sum(tips['tip']>8)
```

```python
##JointPlots
sns.jointplot(data=tips,x="total_bill",y="tip")
```

#here is the only 1 person who have total bill as more than 50$ and that hav only given more than 10 $ as a tip

```python
sns.jointplot(data=tips,x="total_bill",y="tip",hue="smoker")
```

```python
sns.jointplot(data=tips,x="total_bill",y="tip",hue="sex")
```

#customers which are male are giving more bills 
#females give less tips as compared to male through this interpretation

```python
sns.jointplot(data=tips,x="total_bill",y="tip",hue="time")
```
